```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-10 ,10)
y = (x+3)**2
print(x , y)
```

```python
plt.style.use("seaborn")
plt.plot(x,y)
plt.xlabel('x value')
plt.ylabel('F(x)')
plt.show()
```

```python
cur_x = 2
rate = 0.01
precision = 0.000001
prev_step_size = 1
max_iter = 10000
iters = 0
df = lambda x: 2*(x+3)
```

```python
plt.plot(x,y)
while prev_step_size > precision and iters<max_iter:
    pre_x = cur_x
    cur_x = cur_x - rate*df(pre_x)
    prev_step_size = abs(cur_x - pre_x)
    iters = iters+1
    Y = (cur_x+3)**2
    plt.scatter(cur_x , Y)
    print("Iteration",iters,"\nX value is",cur_x)
```

```python
print("The local minimum occurs at", cur_x)
```
